# **Implementing the linear regression**

1. Import neccessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

2. Generate random Dataset.

fetches the california housing dataset and seperates features(x) and target(y)

```python
np.random.seed(42)
x = np.random.rand(50, 1)*100
y = 3.5*x + np.random.randn(50, 1)*20
```

3. Creating and training linear regression model

```python
model = LinearRegression()
model.fit(x, y)
```

4. Predicting y value

```python
y_pred = model.predict(x)
y_pred
```

5. visualising the regression line

```python
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, y_pred, color='red', linewidth=2, label='Regression line')
plt.title('Linear Regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

6. Slope and Intercept

```python
print("Slope(coefficient): ", model.coef_[0][0])
print('Intercept: ', model.intercept_[0])
```
