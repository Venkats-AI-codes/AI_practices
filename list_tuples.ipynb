# **Pandas DataFrame using list of tuples**

1. using pd.DataFrame()

```python
import pandas as pd
data = [('ANSH', 22, 9),
        ('SAHIL', 22, 6),
        ('JAYAN', 23, 8),
        ('AYUSHI', 21, 7),
        ('SPARSH', 20, 8) ]
df = pd.DataFrame(data, columns=['Name', 'age', 'Score'])
df
```

2. using pd.DataFrame.from_records()

```python
df = pd.DataFrame.from_records(data, columns=['Name', 'age', 'Score'])
df
```

3. using pivot()

```python
a = df.pivot(index='Name', columns='Score', values='age')
a
```

some native method to create list of neasted dictionaries

```python
li = [{
        "Student": [{"Exam": 90, "Grade": "a"},
                    {"Exam": 99, "Grade": "b"},
                    {"Exam": 97, "Grade": "c"},
                    ],
        "Name": "Paras Jain"
        },
        {
        "Student": [{"Exam": 89, "Grade": "a"},
                    {"Exam": 80, "Grade": "b"}
                    ],
        "Name": "Chunky Pandey"
        }
        ]
li
```

```python
rows = []

for data in li:
  data_row = data['Student']
  time = data['Name']

  for row in data_row:
    row['Name'] = time
    rows.append(row)
df = pd.DataFrame(rows)
df
```

```python
type(df['Exam'])
```

```python
df = df.pivot_table(index='Name', columns=['Grade'],
                    values=['Exam']).reset_index()
df.columns = ['Name', 'Maths', 'Physics', 'Chemistry']
df
```
