# Plotting from Pandas DataFrame

```python
import pandas as pd
import numpy as np
```

```python
car_sales = pd.read_csv('car_sales.csv')
car_sales.head()
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range("1/1/2026", periods=1000))
ts
```

```python
ts = ts.cumsum()
ts.plot()
```

```python
car_sales['Sale Date'] = pd.date_range('1/1/2026', periods=len(car_sales))
car_sales['Sale Date']
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales.head(10)
```

```python
car_sales.plot(x='Sale Date', y='Total Sales')
```

```python
car_sales.plot(x='Mileage', y='Price', kind='scatter');
```
