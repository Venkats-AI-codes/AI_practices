# **DataFrame using list of lists**

1. using pd.DataFrame function

```python
import pandas as pd
data = [['Venky', 10], ['Venkat', 15], ['Venkateswarlu', 20]]
columns = ['Name', 'age']
df = pd.DataFrame(data, columns=columns)
df
```

2. Handling missing values
The missing values replaced with NaN.

```python
import numpy as np
data = [['Venkat', 28, 'Engineer'],
        ['Venky', None, 'Data Scientist'],
        ['Venkateswarlu', 32, None]]
columns = ['Name', 'age', 'Occupation']

df = pd.DataFrame(data, columns=columns)
df = df.replace({None: np.nan})
df
```

3. DataFrame with different data types


```python
data = [['Venkat', 28, 'Engineer'],
        ['Venky', 25, 'Data Scientist'],
        ['Venkateswarlu', '32', 'Manager']]

df = pd.DataFrame(data, columns=columns)
df['age'] = pd.to_numeric(df['age'])
df
```
