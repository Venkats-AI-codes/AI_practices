Suppose the IQ in a certain population is normally distributed with a mean of mue=100 and standard deviation is 15.
A researcher wants to know if a new drug effects IQ levels, so he recruites 20 patients to try it and records thier IQ levels.
The following code shows how to perform a one sample z-test in python to determine if the new drug causes a significant difference in IQ levels

```python
from statmodels.stats.weightstats import ztest as ztest
data = [88, 92, 94, 94, 96, 97, 97, 97, 99, 99, 106, 109, 109, 109, 110, 112, 112, 113, 114, 115]
ztest(data, value=100)
```

```python
import numpy as np
from scipy.stats import norm
data = data = [88, 92, 94, 94, 96, 97, 97, 97, 99, 99, 106, 109, 109, 109, 110, 112, 112, 113, 114, 115]

mu = 100

alpha = 0.05
n = len(data)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
round(sample_mean, 2)
```

```python
round(sample_std, 2)
```

```python
z_score = (sample_mean - mu)/(sample_std/np.sqrt(n))
round(z_score, 4)
```

```python
p_value = 2*(1-norm.cdf(abs(z_score)))
round(p_value, 4)
```

```python
if p_value<alpha:
  print("Decision: Reject null hypothesis")
else:
  print("Decision: Accept Null hypothesis")
```
