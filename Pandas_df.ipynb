Using pd.DataFrame()

```python
import pandas as pd
lst = [['Venky', 25], ['Ve', 30], ['Venk', 26], ['Venkateswarlu', 22]]
df = pd.DataFrame(lst, columns=['Tag', 'Number'])
df
```

Using pd.DataFrame.from_records()
It is mainly for structured or record oriented data
i.e the inner data of list is treated as records.

```python
data = [['Venky', 28, 'Analyst'], ['Venkateswarlu', 35, 'Manager'], ['Venkat', 29, 'Developer']]
columns = ['Name', 'Age', 'Occupation']
df1 = pd.DataFrame.from_records(data=data, columns=columns)
df1
```

using pd.DataFrame.from_dict()
It creates DataFrame from dictionary from which keys becomes column names and values becomes column data.
In the below example 2D list are transposed using zip(*data) to convert rows into columns before converting into dictionary.

```python
df2 = pd.DataFrame.from_dict(dict(zip(columns, zip(*data))))
df2
```

DataFrame with specified data types
This explicitly defines the columns names and controls the structure of data frame.

```python
data = [['Venky', 'Vana', 25], ['Venkat', 'Vanapa', 30], ['Venkateswarlu', 'Vanaparthi', 26], ['Vvr', 'vanap', 22]]
columns = ['FName', 'LName', 'Age']
df3 = pd.DataFrame(data, columns=columns)
df3
```
